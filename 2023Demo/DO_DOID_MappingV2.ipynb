```python
dong_mapping_df
```

```python

# "동" 값을 ID로 변환
grouped["ID"] = grouped["동"].apply(lambda dongs: [dong_mapping_df[dong_mapping_df['Dong'] == dong]['ID'].values[0] for dong in dongs])
```

```python
# "동" 칼럼을 ID로 변환
df["동_ID"] = df["동"].map(dong_id_mapping)

print(df)
```

```python
import pandas as pd
```

```python
# 데이터 불러오기
rawData = pd.read_csv('result_RelationAnalysis.csv', encoding='utf-8', sep=",")
rawData.head()
```

```python
# A 와 B를 float가 아닌 int로 변경

rawData['A']= rawData['A'].astype(int)
rawData['B']= rawData['B'].astype(int)

# A 와 B를 오름차순으로 결정
rawData = rawData.sort_values(by=['A', 'B'], ascending=[True, True])




```

```python
# rawData
```

```python
# rawData.head()
```

```python
# A,B B,A 정렬
```

```python
# A에서 B로 이동하는 데이터 복사
output1 = rawData.copy()
```

```python

output2 = rawData.copy()
# 열 이름 바꾸기
output2.columns = ['B', 'A', 'similarity']

output2 = output2[[ 'A','B', 'similarity']]
```

```python
output2
```

```python

# 두 데이터프레임 병합
merged_df = pd.concat([output1, output2])

merged_df
```

```python

# # Location_1 및 Location_2 열의 값을 ID로 바꾸기
# merged_df['DB'] = merged_df['DB'].map(region_id_mapping)
# merged_df['OB'] = merged_df['OB'].map(region_id_mapping)

# (Location_1=지역1 & Location_2=지역3)와 (Location_1=지역3 & Location_2=지역1)을 같은 것으로 처리
merged_df['Unique_A'] = merged_df[['A','B']].min(axis=1)
merged_df['Unique_B'] = merged_df[['A','B']].max(axis=1)

# 중복된 데이터 제거
merged_df = merged_df.drop_duplicates(subset=['Unique_A', 'Unique_B'])

# # 불필요한 열 제거
# result_df = result_df.drop(columns=['Unique_DB', 'Unique_OB'])
```

```python
merged_df
```

```python
# 불필요한 열 제거
merged_df = merged_df.drop(columns=['A','B'])

```

```python
merged_df
```

```python
merged_df = merged_df[[ 'Unique_A','Unique_B', 'similarity']]


merged_df
```

```python
merged_df=merged_df.rename(columns={"Unique_A": 'A'})

merged_df=merged_df.rename(columns={"Unique_B": 'B'})


merged_df
```

```python
merged_df = merged_df.sort_values(by=['A', 'B'], ascending=[True, True])
```
